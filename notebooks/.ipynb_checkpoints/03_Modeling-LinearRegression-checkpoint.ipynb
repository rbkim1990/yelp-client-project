### Set up environment.

```python
# Import needed packages.
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression, LassoCV, RidgeCV
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.preprocessing import StandardScaler

# import statsmodels.api as sm

# View plot within notebook.
%matplotlib inline

# Increase number of columns that can be viewed in notebook.
pd.set_option('display.max_columns', 500)
```

### Read in the zip code data file.

```python
# Locate the file.
zip_csv = './data/composite_zip_WI.csv'             

# Load the data.  
zip_df = pd.read_csv(zip_csv,dtype={'zipcode': 'str'})

# Take a look at the data.
zip_df.head()
```

### Model using LinearRegression.

```python
# Select features to use in the model.  These features were selected during the EDA 
# and Feature Engineering process.

# Model with incomplete (from this morning) and remove 8 and 9 businesses.

features = [
            'n_d3_grocery',
            'p_ge_d2_gourmet',
            'n_d3_french',
            'n_d3_coffee',
            'n_s3plus_gourmet',
            'n_s2plus_grocery',
           ]

target   = ['average_agi']
```

```python
# Set features matrix.
X = zip_df[features]

# Set target vector.
y = zip_df[target]
```

```python
# Split the data into train and test sets.
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3, 
                                                    random_state = 42)

# Look at split.
print('X_train:', X_train.shape)
print('y_train:', y_train.shape)
print('X_test :', X_test.shape)
print('y_test :', y_test.shape)
```

```python
# Transform the data to a standard scale.

# Instantiate model.
ss = StandardScaler()

# Transform training and test sets.
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
# Instantiate model.
lr = LinearRegression()

# Perform K-fold cross validation using K=7.
cv_scores = cross_val_score(lr, X_train, y_train, cv=7)

# Look at the scores.
cv_scores
```

```python
# Fit the model.
lr.fit(X_train, y_train)
```

```python
# Predict target in test set.
pred = lr.predict(X_test)
```

### Evaluate LinearRegression Model.

```python
# Score on training data.
lr.score(X_train, y_train)
```

```python
# Score on test data.
lr.score(X_test, y_test)
```

```python
# Check the average CV score.
print('Cross val score mean: ', cv_scores.mean())

# Generate the r2 score.
print('R^2 score           : ', r2_score(y_test, pred))
```

```python
# Calculate the RMSE.
round(np.sqrt(mean_squared_error(y_test, pred)),4)
```

### Visualize LinearRegression Model.

```python
# Set figure.
plt.figure(figsize=(8,8))

# Set up scatterplot.
plt.scatter(pred, y_test, s=20, c='b', zorder=10)

# Set labels and title.
plt.xlabel("Predicted Values")
plt.ylabel("Actual Values")
plt.title("LinearRegression Model")

plt.show()

# Print statistics.
print("R^2: ", r2_score(y_test, pred))
print('Intercept:', lr.intercept_)
print('Coef(s):', lr.coef_)

plt.savefig('.asdja', bbox_inches='tight')
```

### Evaluate Ridge and Lasso models.

Neither the Ridge nor the Lasso model indicated improvement for the general LinearRegression model.

```python
# Evaluate the model.
ridge = RidgeCV(alphas=np.linspace(.1, 10, 100))

# Generate scores.
ridge_scores = cross_val_score(ridge, X_train, y_train, cv=3)

print(ridge_scores)

# Find the mean score.
ridge_scores.mean()
```

```python
# Evaluate the model.
lasso = LassoCV(n_alphas=300)

# Generate scores.
lasso_scores = cross_val_score(lasso, X_train, y_train['average_agi'], cv=3)

print(lasso_scores)

# Find the mean score.
lasso_scores.mean()
```
