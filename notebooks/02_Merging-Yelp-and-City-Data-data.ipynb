# Combine data from Yelp API and city-data.com

```python
# Importing Pandas
import pandas as pd
```

```python
# Reading in the city-data.com data
city = pd.read_csv('../data/zip_city.csv')
city = city.sort_values('zipcode')
city.head()
```

```python
# Reading in the summarized Yelp API data
yelp = pd.read_csv('../data/all_wi_yelp.csv')
yelp = yelp.sort_values('zipcode')
yelp.head()
```

```python
# Merging and saving the data
final_df = city.merge(yelp)
final_df.to_csv('../data/yelp_and_city_data.csv', index=False)
```
